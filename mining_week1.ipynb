```python
from requests_html import HTMLSession

session = HTMLSession()

r = session.get("https://www.liepin.com/zhaopin/?key=数据分析")

# 通过xpath找 “数据分析”岗位的招聘及招聘网址
news = r.html.xpath('//div[@class="job-info"]/h3/a')

for new in news:
    print(new.text)  # 获得工作标题
    print(new.absolute_links)  # 获得工作链接
```

```python
# 引入resquests-html模块
from requests_html import HTMLSession

session = HTMLSession()

# 获取（'某网站'）的数据
r = session.get("https://www.liepin.com/zhaopin/?key=投资经理")

# 通过xpath找 有关“投资经理”岗位的其他信息（薪资、公司地址、要求、年限等）
news = r.html.xpath('//div[@class="job-info"]/h3|//div[@class="job-info"]/p[1]')

# 导出数据（循环）
for new in news:
    print(new.text)  # 获得“投资经理”岗位的其他信息（薪资、地址等）
```

```python
from requests_html import HTMLSession

session = HTMLSession()

r = session.get("http://www.gov.cn/xinwen/tupian/index.htm")

# 通过xpath找 新闻图片图说、图片地址
news = r.html.xpath('//div[@class="subtitle"]/a')
items = r.html.xpath('//div[@class="pannel-image"]/a/img/@src')

# 获取新闻图片信息（循环）
for new in news:
    print("图片信息:",new.text)

# 获取新闻图片地址（循环）
for item in items:
    print("图片地址:",item)
```
